# Mutiple regression exploration

```python
import statsmodels.api as sm # Note the short form
import statsmodels.formula.api as smf # Note the short form
```

```python
estimates_overall = smf.ols('mpg ~ cylinders + horsepower + acceleration + year', auto).fit()
estimates_overall.summary().tables[0]
```

## F-stat Interpretation
F-stat >>1 provides compelling evidence against the null hypothesis H0. In other words, the large F-statistic suggests that at least one of the variables must be related to mpg

```python
estimates_overall.summary().tables[1]
```

## T-stat + CF Interpretation

ALl predictors used have p-values less than 0.05

The confidence intervals are far from zero

## Studentized Residuals vs Leverage

```python
fig, ax = plt.subplots(figsize=(12,8))
fig = sm.graphics.influence_plot(estimates_overall, ax=ax, criterion="cooks")
```

### Residual + Leverage Interpretation
- Studentized Residuals > 3 = outlier
- Leverage ~1 = High leverage

## Testing polynomial features

```python
X = auto.drop('mpg',axis=1)
y = np.log(auto['mpg'])
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, train_size=0.7)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.pipeline import Pipeline
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
```

```python
degrees = [1,2,3]
for i in degrees:
    scaler = StandardScaler()
    poly = PolynomialFeatures(degree=i, interaction_only=False, include_bias = False)
    pipeline = Pipeline([('scale', scaler),
                     ('poly', poly), 
                     ('model', LinearRegression())])
    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    print('MSE: ' + str(mse), 'R2 ' + str(r2),sep='\n')
```

```python
scaler = StandardScaler()
poly = PolynomialFeatures(degree=2, interaction_only=False, include_bias = False)
pipeline = Pipeline([('scale', scaler),
         ('poly', poly), 
         ('model', LinearRegression())])
pipeline.fit(X_train, y_train)
y_pred = pipeline.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(mse,r2, sep='\n')
```

```python
coeff = pipeline.named_steps['model'].coef_
intercept = pipeline.named_steps['model'].intercept_
```

```python
print(coeff,intercept,sep='\n')
```

### Interaction terms

```python
polynomialfit = smf.ols(
    formula=
    'mpg ~ cylinders * displacement + cylinders*horsepower + cylinders*weight + cylinders*acceleration + acceleration*year + displacement*horsepower + displacement*weight + displacement*acceleration + displacement*year +  horsepower*weight + horsepower*acceleration + horsepower*year + weight*acceleration + weight*year + acceleration*year',
    data=auto).fit()
```

```python
polynomialfit.summary().tables[0]
```

#### Interpretation
From p-values, we need to include displacement, weight, displacement:weight, acceleration, year, acceleration:year, cylinder, cylinder:displacement

```python
polynomialfit.pvalues[polynomialfit.pvalues < 0.05].sort_values()
```

```python
polynomialfit1 = smf.ols(
    formula=
    'mpg ~ cylinders * displacement + acceleration*year +  displacement*weight',
    data=auto).fit()
```

### Polynomial terms

```python
models_d = [smf.ols(formula='mpg ~ {0} + np.power({0}, 2) + np.power({0}, 3)'.format(f), data=auto).fit()
            for f in X_train.columns[:]]

for model in models_d:
    dp = 4
    display(pd.DataFrame({'coef':    np.around(model.params, dp),
                          'std_err': np.around(model.bse, dp),
                          't':       np.around(model.params / model.bse, dp),
                          'P>|t|':   np.around(model.pvalues, dp),
                          '[0.025':  np.around(model.conf_int(alpha=0.05)[0], dp),
                          '0.975]':  np.around(model.conf_int(alpha=0.05)[1], dp)}))
```

```python
sig = pd.concat([model.pvalues[model.pvalues < 0.05] for model in models_d])
pd.DataFrame({'P>|t|':sig.drop('Intercept')})
```

```python
X_significant = X
X_significant.head()
```

```python
X_significant['(cylinders, 2)'] = X_significant['cylinders']**2
X_significant['(cylinders, 3)'] = X_significant['cylinders']**3
X_significant['(displacement, 2)'] = X_significant['displacement']**2
X_significant['(horsepower, 2)'] = X_significant['horsepower']**2
X_significant['displacement:weight'] = X_significant['displacement']*X_significant['weight']
X_significant['displacement:cylinders'] = X_significant['displacement']*X_significant['cylinders']
X_significant['acceleration:year'] = X_significant['acceleration']*X_significant['year']

X_significant.head()
```

```python
from sklearn.model_selection import train_test_split

X_train_sig, X_test_sig, y_train_sig, y_test_sig = train_test_split(X_significant, y, random_state=0, train_size=0.7)
```

```python
scaler = StandardScaler()
pipeline = Pipeline([('scale', scaler),
         ('model', LinearRegression())])

pipeline.fit(X_train_sig, y_train_sig)
y_pred_sig = pipeline.predict(X_test_sig)
mse_sig  = mean_squared_error(y_test_sig , y_pred_sig )
r2_sig  = r2_score(y_test_sig , y_pred_sig)
```

```python
print(mse_sig,r2_sig,sep='\n')
```

## Conclusion
We tried 2 approaches to adding interaction and polynomial terms.

The first approach was a blanket approach of including every interaction and polynomial term without statistical consideration using sklearn, and cross-validating to find the optimal degree.

The second approach took care of examining the individual p-values of every interaction and polynomial term to be added.

The results are quite similar, although the statistical consideration approach would make more sense by fitting EDA better.
